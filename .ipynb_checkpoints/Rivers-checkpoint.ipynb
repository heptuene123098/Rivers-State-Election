```python
# Load the Rivers
Rivers = pd.read_excel("Rivers.xlsx")

# Extract latitude and longitude as a numpy array
coordinate = Rivers[['Latitude', 'Longitude']].values

# Define the proximity radius (in kilometers)
epsilon = 1.0 / 6371.0  # 1 km in radians (earth radius = 6371 km)

# Perform DBSCAN clustering
db = DBSCAN(eps=epsilon, min_samples=1, algorithm='ball_tree', metric='haversine').fit(np.radians(coordinate))

# Add cluster labels to the Rivers
Rivers['Cluster'] = db.labels_

# Save the results to a new CSV file
output_path = 'clustered_polling_units.csv'
Rivers.to_csv(output_path, index=False)

print("Clustering completed and results saved to", output_path)
```

```python
import folium

# Create a base map
map_center = [Rivers['Latitude'].mean(), Rivers['Longitude'].mean()]
map = folium.Map(location=map_center, zoom_start=10)

# Add points to the map
for _, row in Rivers.iterrows():
    folium.CircleMarker(
        location=(row['Latitude'], row['Longitude']),
        radius=5,
        popup=f"Cluster: {row['Cluster']}",
        color='blue' if row['Cluster'] != -1 else 'red',  # Red for noise points
        fill=True,
        fill_color='blue' if row['Cluster'] != -1 else 'red'
    ).add_to(map)

# Save the map to an HTML file
map.save('polling_units_clusters_map.html')

print("Map created and saved to polling_units_clusters_map.html")
```

```python
# Analyze clusters
cluster_analysis = Rivers['Cluster'].value_counts().reset_index()
cluster_analysis.columns = ['Cluster', 'Number of Polling Units']

# Add cluster_analysis labels to the Rivers
Rivers['Cluster_analysis'] = db.labels_

# Save the cluster analysis results to a new CSV file
analysis_output_path = 'cluster_analysis.csv'
Rivers.to_csv(output_path, index=False)

print("Cluster analysis completed and results saved to", analysis_output_path)
```

```python
# Calculate outlier scores for each party within each cluster
parties = ['APC', 'LP', 'PDP', 'NNPP']
outlier_scores = []

for cluster_label in Rivers['Cluster'].unique():
    if cluster_label == -1:
        continue  # Skip noise points
    
    cluster_Rivers = Rivers[Rivers['Cluster'] == cluster_label].copy()
    
    for party in parties:
        votes = cluster_Rivers[party]
        mean_votes = votes.mean()
        std_votes = votes.std()

        # Calculate the Z-score for each polling unit in the cluster
        if std_votes != 0:
            z_scores = (votes - mean_votes) / std_votes
        else:
            z_scores = [0] * len(votes)  # If std is zero, all values are identical, and z-score is 0

        cluster_Rivers.loc[:, f'{party}_Outlier_Score'] = z_scores

    outlier_scores.append(cluster_Rivers)

# Concatenate the results
outlier_scores_df = pd.concat(outlier_scores)

# Save the outlier scores to a new CSV file
outlier_scores_output_path = 'polling_units_with_outlier_scores_1.csv'
outlier_scores_df.to_csv(outlier_scores_output_path, index=False)

print("Outlier scores calculated and results saved to", outlier_scores_output_path)

```

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
from sklearn.ensemble import IsolationForest

# Load the dataset with outlier scores
outlier_scores_path = 'polling_units_with_outlier_scores_2.csv'
outlier_scores_df = pd.read_csv(outlier_scores_path)

# Define parties
parties = ['APC', 'LP', 'PDP', 'NNPP']

# Dictionary to store top outliers and their closest polling units
top_outliers = {party: [] for party in parties}

# Function to calculate distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).kilometers

# Sort by outlier scores for each party
for party in parties:
    # Sort by party outlier score descending
    sorted_df = outlier_scores_df.sort_values(by=f'{party}_Outlier_Score', ascending=False)
    
    # Select top 3 outliers
    top_outliers_df = sorted_df.head(3)
    
    # Extract closest polling units for each top outlier
    for index, row in top_outliers_df.iterrows():
        # Calculate distances to all other polling units in the same cluster
        distances = []
        for _, other_row in outlier_scores_df.iterrows():
            if other_row['Cluster'] == row['Cluster'] and (other_row['Latitude'] != row['Latitude'] or other_row['Longitude'] != row['Longitude']):
                distance = calculate_distance(row['Latitude'], row['Longitude'], other_row['Latitude'], other_row['Longitude'])
                distances.append((distance, other_row['State'], other_row['LGA'], other_row['Ward'], other_row['PU-Code'], other_row['Full Address']))
        
        # Sort distances and get the closest polling units
        closest_polling_units = sorted(distances, key=lambda x: x[0])[:3]
        
        top_outliers[party].append({
            'Polling_Unit': row['PU-Code'],
            'Outlier_Score': row[f'{party}_Outlier_Score'],
            'Closest_Polling_Units': closest_polling_units
        })

# Print or report findings
for party in parties:
    print(f"Top 3 outliers for {party}:")
    for i, outlier in enumerate(top_outliers[party]):
        print(f"Outlier {i+1}:")
        print(f"  Polling Unit: {outlier['Polling_Unit']}")
        print(f"  Outlier Score: {outlier['Outlier_Score']}")
        print("  Closest Polling Units:")
        for unit in outlier['Closest_Polling_Units']:
            print(f"    - State: {unit[1]}, LGA: {unit[2]}, Ward: {unit[3]}, PU-Code: {unit[4]}, Address: {unit[5]}, Distance: {unit[0]:.2f} km")
        print()

# Save mean values to CSV
mean_values = []

for cluster_label in outlier_scores_df['Cluster'].unique():
    if cluster_label == -1:
        continue  # Skip noise points
    
    cluster_data = outlier_scores_df[outlier_scores_df['Cluster'] == cluster_label]
    
    cluster_mean_values = {'Cluster': cluster_label}
    for party in parties:
        cluster_mean_values[f'{party}_Mean'] = cluster_data[party].mean()
        cluster_mean_values[f'{party}_Std'] = cluster_data[party].std()
    
    mean_values.append(cluster_mean_values)

mean_df = pd.DataFrame(mean_values)
mean_df.to_csv('cluster_mean_values.csv', index=False)

print("Mean values saved to cluster_mean_values.csv")
```

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
from sklearn.ensemble import IsolationForest

# Load the dataset with outlier scores
outlier_scores_path = 'polling_units_with_outlier_scores_2.csv'
outlier_scores_df = pd.read_csv(outlier_scores_path)

# Define parties
parties = ['APC', 'LP', 'PDP', 'NNPP']

# Dictionary to store top outliers and their closest polling units
top_outliers = {party: [] for party in parties}

# Function to calculate distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).kilometers

# DataFrame to store top outliers and their closest polling units
outliers_df = pd.DataFrame(columns=[
    'Party', 'Outlier_Polling_Unit', 'Outlier_Score', 
    'Closest_Polling_Unit_1', 'Distance_1', 'Closest_Polling_Unit_2', 'Distance_2', 'Closest_Polling_Unit_3', 'Distance_3',
    'State_1', 'LGA_1', 'Ward_1', 'Address_1', 'State_2', 'LGA_2', 'Ward_2', 'Address_2', 'State_3', 'LGA_3', 'Ward_3', 'Address_3'
])

# Sort by outlier scores for each party
for party in parties:
    # Sort by party outlier score descending
    sorted_df = outlier_scores_df.sort_values(by=f'{party}_Outlier_Score', ascending=False)
    
    # Select top 3 outliers
    top_outliers_df = sorted_df.head(3)
    
    # Extract closest polling units for each top outlier
    for index, row in top_outliers_df.iterrows():
        # Calculate distances to all other polling units in the same cluster
        distances = []
        for _, other_row in outlier_scores_df.iterrows():
            if other_row['Cluster'] == row['Cluster'] and (other_row['Latitude'] != row['Latitude'] or other_row['Longitude'] != row['Longitude']):
                distance = calculate_distance(row['Latitude'], row['Longitude'], other_row['Latitude'], other_row['Longitude'])
                distances.append((distance, other_row['State'], other_row['LGA'], other_row['Ward'], other_row['PU-Code'], other_row['Full Address']))
        
        # Sort distances and get the closest polling units
        closest_polling_units = sorted(distances, key=lambda x: x[0])[:3]
        
        # Create a record for this outlier and its closest polling units
        record = {
            'Party': party,
            'Outlier_Polling_Unit': row['PU-Code'],
            'Outlier_Score': row[f'{party}_Outlier_Score']
        }
        for i, unit in enumerate(closest_polling_units):
            record[f'Closest_Polling_Unit_{i+1}'] = unit[4]
            record[f'Distance_{i+1}'] = unit[0]
            record[f'State_{i+1}'] = unit[1]
            record[f'LGA_{i+1}'] = unit[2]
            record[f'Ward_{i+1}'] = unit[3]
            record[f'Address_{i+1}'] = unit[5]
        
        # Convert the record to a DataFrame and concatenate to outliers_df
        outliers_df = pd.concat([outliers_df, pd.DataFrame([record])], ignore_index=True)

# Save the outliers to a CSV file
outliers_df.to_csv('top_outliers_with_closest_units.csv', index=False)

print("Top outliers and their closest polling units saved to top_outliers_with_closest_units.csv")

# Save mean values to CSV
mean_values = []

for cluster_label in outlier_scores_df['Cluster'].unique():
    if cluster_label == -1:
        continue  # Skip noise points
    
    cluster_data = outlier_scores_df[outlier_scores_df['Cluster'] == cluster_label]
    
    cluster_mean_values = {'Cluster': cluster_label}
    for party in parties:
        cluster_mean_values[f'{party}_Mean'] = cluster_data[party].mean()
        cluster_mean_values[f'{party}_Std'] = cluster_data[party].std()
    
    mean_values.append(cluster_mean_values)

mean_df = pd.DataFrame(mean_values)
mean_df.to_csv('cluster_mean_values.csv', index=False)

print("Mean values saved to cluster_mean_values.csv")
```

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
import folium

# Load the dataset with outlier scores
outlier_scores_path = 'polling_units_with_outlier_scores_2.csv'
outlier_scores_df = pd.read_csv(outlier_scores_path)

# Define parties
parties = ['APC', 'LP', 'PDP', 'NNPP']

# Dictionary to store top outliers and their closest polling units
top_outliers = {party: [] for party in parties}

# Function to calculate distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).kilometers

# Sort by outlier scores for each party
for party in parties:
    # Sort by party outlier score descending
    sorted_df = outlier_scores_df.sort_values(by=f'{party}_Outlier_Score', ascending=False)
    
    # Select top 3 outliers
    top_outliers_df = sorted_df.head(3)
    
    # Extract closest polling units for each top outlier
    for index, row in top_outliers_df.iterrows():
        # Calculate distances to all other polling units in the same cluster
        distances = []
        for _, other_row in outlier_scores_df.iterrows():
            if other_row['Cluster'] == row['Cluster'] and (other_row['Latitude'] != row['Latitude'] or other_row['Longitude'] != row['Longitude']):
                distance = calculate_distance(row['Latitude'], row['Longitude'], other_row['Latitude'], other_row['Longitude'])
                distances.append((distance, other_row['Latitude'], other_row['Longitude'], other_row['State'], other_row['LGA'], other_row['Ward'], other_row['PU-Code'], other_row['Full Address']))
        
        # Sort distances and get the closest polling units
        closest_polling_units = sorted(distances, key=lambda x: x[0])[:3]
        
        top_outliers[party].append({
            'Polling_Unit': row['PU-Code'],
            'Outlier_Score': row[f'{party}_Outlier_Score'],
            'Latitude': row['Latitude'],
            'Longitude': row['Longitude'],
            'Closest_Polling_Units': closest_polling_units
        })

# Prepare the data for saving to a CSV
outliers_data = []

for party in parties:
    for outlier in top_outliers[party]:
        record = {
            'Party': party,
            'Outlier_Polling_Unit': outlier['Polling_Unit'],
            'Outlier_Score': outlier['Outlier_Score'],
            'Latitude': outlier['Latitude'],
            'Longitude': outlier['Longitude']
        }
        
        for i, unit in enumerate(outlier['Closest_Polling_Units']):
            record[f'Closest_Polling_Unit_{i+1}'] = unit[3]
            record[f'Distance_{i+1}'] = unit[0]
            record[f'Latitude_{i+1}'] = unit[1]
            record[f'Longitude_{i+1}'] = unit[2]
        
        outliers_data.append(record)

outliers_df = pd.DataFrame(outliers_data)
outliers_df.to_csv('top_outliers_with_closest_units.csv', index=False)

# Generate the interactive map
map_center = [9.0820, 8.6753]
outlier_map = folium.Map(location=map_center, zoom_start=6)

# Add outliers and their closest polling units to the map
for index, row in outliers_df.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        # Add outlier marker
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"{row['Party']} Outlier: {row['Outlier_Polling_Unit']} (Score: {row['Outlier_Score']})",
            icon=folium.Icon(color='red', icon='exclamation-sign')
        ).add_to(outlier_map)
    
        # Add closest polling unit markers
        for i in range(1, 4):
            if pd.notna(row[f'Latitude_{i}']) and pd.notna(row[f'Longitude_{i}']):
                folium.Marker(
                    location=[row[f'Latitude_{i}'], row[f'Longitude_{i}']],
                    popup=f"Closest Unit {i}: {row[f'Closest_Polling_Unit_{i}']} (Distance: {row[f'Distance_{i}']:.2f} km)",
                    icon=folium.Icon(color='blue', icon='info-sign')
                ).add_to(outlier_map)

# Save the map to an HTML file
outlier_map.save('top_outliers_map.html')

print("Map saved to top_outliers_map.html")
```
